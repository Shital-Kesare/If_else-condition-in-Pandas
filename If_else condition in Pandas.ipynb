```python
import pandas as pd

df = pd.read_excel('orders.csv.xlsx')
df
```

```python
df[df['city'] == 'Banglore']
```

```python
df.loc[df['city'] == 'Banglore','state'] = 'Karnataka'
df.loc[df['city'] == 'Chennai','state'] = 'TN'
df.loc[df['city'] == 'Mysore','state'] = 'Karnataka'
```

```python
df.loc[df['city'].isin(['Banglore','Mysore']),'state'] = 'Karnataka' 
df.loc[df['city'].isin(['Chennai']),'state'] = 'Tamilnadu'
```

```python
df
```

```python
df.loc[(df['profit']<=250),'profit_cat'] = 'Low Profit'
df.loc[(df['profit']>250) & (df['profit']>=500),'profit_cat'] = 'Medium Profit'
df.loc[df['profit']>=500,'profit_cat'] = 'High Profit'
```
